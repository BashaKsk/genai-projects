### Streaming

**Streaming** = receiving the model's response *as it's generated*, token by token, instead of waiting for the full reply.

- `model.invoke(...)` → blocks, returns one full `AIMessage` at the end.
- `model.stream(...)` → yields `AIMessageChunk` objects as the model produces text (the typewriter effect in ChatGPT/Claude).

**Pattern:**

```python
for chunk in model.stream("Tell me a short story"):
    print(chunk.content, end="", flush=True)
```

Use it for chat UIs, long outputs, and anywhere "feels fast" matters.

````python
### Here we have to wait until model generates the 100 lines

model.invoke("Tell me 100 lines story about naruto")
````

````python
for chunk in model.stream("Tell me 100 lines story about naruto"):
    print(chunk.text, end="|" , flush=True)
````

````python
for chunk in model.stream("why do parrots have colorful feathers??"):
    print(chunk.text, end="" , flush=True)
````

### Batch

**Batch** = sending multiple prompts to the model in one call. `model.batch([...])` takes a list of prompts and returns a list of `AIMessage` responses, running them in parallel under the hood for higher throughput.

```python
responses = model.batch(["Hi", "How are you?", "What's up?"])
```

````python
responses = model.batch(["Hi", "How are you?", "What's up?"])

for response in responses:
    print(response)
````

````python
responses = model.batch(
    ["Hi", "How are you?", "What's up?"],
    config= {
        'max_concurrency' : 5,  # Limit to 5 parallel calls
    }

    )

for response in responses:
    print(response)
````
