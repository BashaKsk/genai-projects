```python
numbers = [1,2,3,4,5,6,7,8]

square = lambda x: x**2

list(map(square, numbers))
```

```python
## lamda function with map
list(map(lambda x: x *x, numbers))
```

```python
### Map multiple iterables

numbers1= [1,2,3]

numbers2= [4,5,6]

added_numbers = list(map(lambda x, y: x+y, numbers1, numbers2))

print(added_numbers)
```

```python
## map to converet a list of strings to integer


str_number = ['1', '2', '3']


list(map(int, str_number))
```

```python
words = ['apple', 'banana', 'cherry']


list(map(str.upper, words))

```

```python
def get_name(person):
    return person['name']


people = [

    {
        'name': 'krish', 'age': 12,
       
    },

    {
         'name': 'basha', 'age': 24
    }
]


print(list(map(get_name, people)))
```
