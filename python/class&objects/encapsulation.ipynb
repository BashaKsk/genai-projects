# Encapsulation

**Encapsulation** is one of the four fundamental principles of Object-Oriented Programming (OOP). It is the practice of bundling data (attributes) and the methods that operate on that data into a single unit — a **class** — while restricting direct access to some of the object's components.

## Key Ideas

- **Data hiding**: Internal state of an object is kept private and protected from outside interference.
- **Controlled access**: Access to the data is provided through public methods (getters and setters), allowing validation and consistency.
- **Modularity**: Each object manages its own state, making code easier to maintain and reason about.

## Access Modifiers in Python

| Modifier | Convention | Meaning |
|----------|------------|---------|
| Public | `name` | Accessible from anywhere |
| Protected | `_name` | Intended for internal use within the class and its subclasses |
| Private | `__name` | Name-mangled; not directly accessible from outside the class |

## Why Use Encapsulation?

1. **Security** — prevents unintended modification of data.
2. **Flexibility** — internal implementation can change without affecting external code.
3. **Maintainability** — clear boundaries between what is exposed and what is hidden.
4. **Validation** — setters can enforce rules before changing the state.


```python
## Encapsulation with GEtter and Setter methods

### Public
### Protected
### Private variables or access modifiers



class Person:
    def __init__(self, name, age):
        self.name = name ## publick variables
        self.age= age ## public  variable


person = Person("Basha", 30)

person.name
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name ## private variables
        self._age= age ## protected  variable can be accesed in derived class
        self.gender = gender ## public

person = Person("Basha", 30, 'male')

dir(person)

#'_Person__age',
#  '_Person__name',
```

```python
class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)


employee1 = Employee("ksk", 24, 'male')

employee1._age
```

```python
## Encapsulaiton with Getter and Setter

class Person:

    def __init__(self, name, age,):
        self.__name = name ## private variables
        self.__age= age ## protected  variable can be accesed in derived class

    ### getter method for name
    def get_name(self):
        return self.__name
    
    # setter mothd
    def set_name(self, name):
        self.__name = name


person = Person("Krish", 34)


## Acess and modify private variabels using getter and setters

print(person.get_name())
```
