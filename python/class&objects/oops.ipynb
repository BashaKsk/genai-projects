# Object-Oriented Programming (OOP)

**Object-Oriented Programming** is a programming paradigm that organizes code around **objects** — bundles of *data* (attributes) and *behavior* (methods) — instead of around functions and procedures.

It's built on four core principles:

1. **Encapsulation** — bundling data and the functions that operate on it into a single unit (a class), and hiding internal state behind a controlled interface.
2. **Abstraction** — exposing only the relevant details and hiding the underlying complexity.
3. **Inheritance** — a class can derive from another class and reuse/extend its behavior.
4. **Polymorphism** — the same method name can behave differently depending on the object that invokes it.

OOP is useful when your problem maps naturally to "things" with state and behavior — users, accounts, requests, files, sessions, etc.

---

## What is a Class?

A **class** is a **blueprint** for creating objects. It defines:
- What **data** every instance will hold (attributes)
- What **operations** can be performed on that data (methods)

An **object** (also called an *instance*) is a concrete value created from a class. One class can produce many independent objects.

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name        # attribute
        self.breed = breed      # attribute

    def bark(self):             # method
        print(f"{self.name} says woof!")

# Creating objects (instances) from the class
dog1 = Dog("Rex", "Labrador")
dog2 = Dog("Buddy", "Beagle")

dog1.bark()   # Rex says woof!
dog2.bark()   # Buddy says woof!
```

**Key terms:**
- `class Dog:` — the class definition (the blueprint)
- `dog1`, `dog2` — instances (real objects in memory)
- `__init__` — the **constructor**, called automatically when you create an instance
- `self` — refers to the current instance; Python passes it implicitly when you call `dog1.bark()`
- `self.name` — an **instance attribute**, unique to each object

````python
## Define a class with instance methods

class Dog:

    ## Constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says wooof")

dog1 = Dog('Buddy', 5)

dog1.bark()

dog2 = Dog('Lucy', 5)

dog2.bark()

````
