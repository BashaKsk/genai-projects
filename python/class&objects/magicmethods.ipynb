# Magic Methods (Dunder Methods)

**Magic methods** — also called **dunder methods** (double underscore) — are special methods in Python that begin and end with `__` (e.g., `__init__`, `__str__`, `__add__`). They let your classes integrate with Python's built-in syntax and operators.

You don't usually call them directly — Python calls them for you when you use a corresponding operator, built-in function, or syntax construct.

## Why They Matter

They allow custom objects to behave like built-in types — supporting `+`, `-`, `len()`, `print()`, `==`, indexing, iteration, and more.

```python
obj1 + obj2          # calls obj1.__add__(obj2)
len(obj)             # calls obj.__len__()
print(obj)           # calls obj.__str__()
obj[0]               # calls obj.__getitem__(0)
```

## Common Magic Methods

### Object Lifecycle
| Method | Triggered by | Purpose |
|--------|--------------|---------|
| `__init__` | `MyClass()` | Initialize a new instance |
| `__new__` | object creation | Actually creates the instance (rarely overridden) |
| `__del__` | `del obj` / garbage collection | Cleanup before destruction |

### String Representation
| Method | Triggered by | Purpose |
|--------|--------------|---------|
| `__str__` | `str(obj)`, `print(obj)` | Human-readable string |
| `__repr__` | `repr(obj)`, REPL | Unambiguous developer string |

### Arithmetic Operators
| Method | Operator |
|--------|----------|
| `__add__` | `+` |
| `__sub__` | `-` |
| `__mul__` | `*` |
| `__truediv__` | `/` |
| `__floordiv__` | `//` |
| `__mod__` | `%` |
| `__pow__` | `**` |

### Comparison Operators
| Method | Operator |
|--------|----------|
| `__eq__` | `==` |
| `__ne__` | `!=` |
| `__lt__` | `<` |
| `__le__` | `<=` |
| `__gt__` | `>` |
| `__ge__` | `>=` |

### Container / Sequence Behavior
| Method | Triggered by |
|--------|--------------|
| `__len__` | `len(obj)` |
| `__getitem__` | `obj[key]` |
| `__setitem__` | `obj[key] = value` |
| `__delitem__` | `del obj[key]` |
| `__contains__` | `x in obj` |
| `__iter__` | `for x in obj` |
| `__next__` | `next(obj)` |

### Callable & Context Managers
| Method | Triggered by |
|--------|--------------|
| `__call__` | `obj()` — makes instance callable like a function |
| `__enter__` / `__exit__` | `with obj:` block |

## Quick Example

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"Book: {self.title}"

    def __len__(self):
        return self.pages

    def __eq__(self, other):
        return self.title == other.title

b = Book("Python 101", 300)
print(b)        # Book: Python 101  -> __str__
print(len(b))   # 300               -> __len__
```

## Key Takeaway

Magic methods are the hooks that make Python "Pythonic" — they let your objects feel like native parts of the language instead of foreign types.
