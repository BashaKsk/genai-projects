```python
##3 Read the content from source text file and write it to a destination text file

with open('../../content/sample.txt', 'r') as source_file:
    content = source_file.read()

with open('../../content/destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum((len(line.split()) for line in lines))
        char_count = (len(line) for line in lines)

        return line_count, word_count, char_count

file_path = '../../content/destination.txt'
lines, words, chars = count_text_file(file_path)

print(f'Lines: {lines}, Words: {words}, chars: {chars}')
```
