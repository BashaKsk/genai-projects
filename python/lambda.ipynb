```python
def addition(a,b):
    return a+b
```

```python
print(addition(1,3))
```

```python
addition = lambda a,b: a+b

type(addition)

print(addition(1,2))
```

```python
def even(num):
    if(num %2 == 0):
        return True
even(24)
```

```python
is_even = lambda num: num % 2==0

is_even(12)
```

```python
numbers = [1,2,3,4,5,6]


def square(number):
    return number **2

square(2)
```

```python
square = map(lambda num: num **2, numbers)

list(square)
```
